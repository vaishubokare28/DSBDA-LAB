## Visualize the data using Python libraries matplotlib, seaborn by plotting the graphs  on Heart Disease Dataset


---

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading csv file
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
# Check info
df.info()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Rename target column if needed
# 1 = disease, 0 = no disease
df['target'] = df['target'].astype('category')
```

# Histogram

```python
plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

# Count Plot

```python
sns.countplot(x='target', data=df)
plt.title("Heart Disease Count (0 = No, 1 = Yes)")
plt.show()
```

# Box Plot

```python
sns.boxplot(x=df['chol'])
plt.title("Cholesterol Distribution")
plt.show()
```

# Scatter Plot

```python
sns.scatterplot(x='age', y='chol', hue='target', data=df)
plt.title("Age vs Cholesterol")
plt.show()
```

# Heatmap

```python
plt.figure(figsize=(10,8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# Line Plot

```python
df.groupby('age')['chol'].mean().plot()
plt.title("Average Cholesterol by Age")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

# Pair Plot

```python
sns.pairplot(df[['age','trestbps','chol','thalach','target']], hue='target')
plt.show()
```
